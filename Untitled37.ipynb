**Q1: What is Simple Linear Regression (SLR)?Explain its purpose.**

Ans: Simple Linear Regression (SLR) is a statistical method use to understand the relationship between one independent variable (X) and one dependent variable (Y).It represents this relationship using a straight-line equation:

    Y=a+bx

where



*  a is the intercept
*  b is the slope of the line

The main purpose of Simple Linear Regression is to predict the value of the dependent variable based on the independent variable and to analyze how changes in X affect Y. It is widely used in data analysis,forecasting,and machine learning.



**Q2. What are the key assumptions of Simple Linear Regression?**

Ans: The key assumptions of Simple Linear Regression are:

1. Linearity: There is a linear relationship between the independent and dependent variables
2. Independence: The observations are independent of each other.

1. Homoscedasticity: The variance of errors is constant across all levels of the independent variable.
2. Normality: The errors terms are normally distributed.

1. No Multicollinearity: Since SLR has only one independent variable,multicollinearity does not exist.

**Q7.Name and briefly describe three comoon evaluation metrics for regression models.**

Ans:

1. Mean Absolute Error (MAE):Average of absolute differences between actual and predicted values.

2. Mean Squared Error (MSE):Average of squared differences between actual and predicted values.

3. R-squared (R2):Shows how well the model explains the variation in the data.










**Q8.What is the purpose of the R-squared metric in regression analysis?**

Ans: R-squared measures how much variation in the dependent variable.Its value lies between 0 and 1,where a higher value indicates a batter model fit.

**Q9.Write Python code to fit a simple linear regression model using scikit-learn and print the slope and intercept.**

```python
from sklearn.linear_model import LinearRegression
import numpy as np

X=np.array([1,2,3,4,5]).reshape(-1,1)
Y=np.array([2,4,6,8,10])

model=LinearRegression()
model.fit(X,Y)

print("Slope:",model.coef_[0])
print("Intercept:",model.intercept_)
```

**Q10.How do you interpret the coefficients in a simple linear regression model?**

Ans: In a simple linear regression model Y=a+bX,the intercept (a) represents the value of Y when X is zero,and the cofficient (b) represents the change in Y for a one-unit increase in X.
